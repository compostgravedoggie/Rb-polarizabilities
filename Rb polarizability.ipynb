```python
import numpy as np
import scipy.optimize as opt
import numpy as np
import pylab as plt
```

```python
S12 = np.genfromtxt('Rb5s-pol.txt',skip_header=1, usecols=[0,1])
P12 = np.genfromtxt('Rb5p1-pol.txt',skip_header=1, usecols=[0,1])
P32 = np.genfromtxt('Rb5p3-pol.txt',skip_header=2, usecols=[0,1,3])
```

```python
plt.figure(figsize=(1.6*6,6))
plt.plot(S12[:,0], S12[:,1], 'C0', label = r'$5s^2S_{1/2}$')
plt.plot(P12[:,0], P12[:,1], 'C3', label = r'$5s^2P_{1/2}$')
plt.plot(P32[:,0], P32[:,1], 'forestgreen', linestyle='--', label = r'$5s^2P_{3/2}$ ($m_J = 1/2$)')
plt.plot(P32[:,0], P32[:,2], 'forestgreen', linestyle=':', label = r'$5s^2P_{3/2}$ ($m_J = 3/2$)')
plt.xlabel('Wavelength [nm]')
plt.ylabel('Polarizability [a.u.]')
plt.xlim([600,1000])
plt.ylim([-5000,5000])
plt.legend()
plt.grid()
```

### Differential polarizabilities

```python
plt.figure(figsize=(1.6*6,6))
plt.plot(P12[:,0], P12[:,1]-S12[:,1], 'C3', label = r'$5s^2P_{1/2}$ - $5s^2S_{1/2}$')
plt.plot(P32[:,0], P32[:,1]-S12[:,1], 'forestgreen', linestyle='--', label = r'$5s^2P_{3/2}$ ($m_J = 1/2$) - $5s^2S_{1/2}$')
plt.plot(P32[:,0], P32[:,2]-S12[:,1], 'forestgreen', linestyle=':', label = r'$5s^2P_{3/2}$ ($m_J = 3/2$) - $5s^2S_{1/2}$')
plt.xlabel('Wavelength [nm]')
plt.ylabel('Polarizability [a.u.]')
plt.xlim([600,1000])
plt.ylim([-5000,5000])
plt.legend()
plt.grid()
```
